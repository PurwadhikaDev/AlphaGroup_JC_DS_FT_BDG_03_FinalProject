[***Loan to Deposit (LTD) Ratio***](https://www.ocbc.id/id/article/2023/04/06/loan-to-deposit-ratio) adalah  rasio yang menggambarkan seberapa mampu sebuah bank untuk membayar kembali dana depositor yang telah digunakan untuk memberikan kredit atau pinjaman kepada nasabah lainnya.

Tingkat LTD di Portugal yang sudah mencapai **157,8%** menjadi sangat berbahaya,  ini berarti bahwa bank telah meminjamkan lebih banyak uang dibandingkan dengan jumlah deposito yang diterima. Terdapat beberapa cara untuk membuat bank menjadi sehat kembali, tapi cara yang paling utama harus dilakukan adalah dengan menarik lebih banyak deposito untuk meningkatkan basis dana.

Model dievaluasi menggunakan metrik evaluasi yang relevan:

- **Presisi (Precission):** Mengukur proporsi prediksi positif yang benar dari total prediksi positif..
- **Recall:** Mengukur proporsi prediksi positif yang benar dari total kasus positif yang sebenarnya. Penting untuk meminimalkan False Negatives.
- **F1-Score:** Menggabungkan presisi dan recall menjadi satu metrik dengan mengambil rata-rata harmonisnya.
- **AUC-ROC:** Mengukur kemampuan model untuk membedakan antara kelas pada berbagai tingkat ambang batas.

Berdasarkan metrik evaluasi, model dengan kinerja terbaik dipilih. Recall biasanya menjadi metrik utama yang dioptimalkan untuk meminimalkan kerugian akibat nasabah potensial yang tidak terdeteksi. Fokus dalam permodelan ini adalah nasabah yang melakukan deposito setelah menerima campaign dari bank, maka target yang digunakan adalah deposit dengan detail sebagai berikut:

**Target:**
- 0 : Tidak melakukan deposito (No Deposit)
- 1 : Melakukan deposito (Deposit)

**Type 1 Error: False Positive**
- **Penjelasan:** Nasabah yang sebenarnya tidak melakukan deposito tetapi diprediksi melakukan deposito.
- **Konsekuensi:** Bank mungkin akan mengalokasikan sumber daya untuk nasabah yang sebenarnya tidak akan mendepositokan dana mereka, mengakibatkan pemborosan sumber daya.

**Type 2 Error: False Negative**
- **Penjelasan:** Nasabah yang sebenarnya melakukan deposito tetapi diprediksi tidak melakukan deposito.
- **Konsekuensi:** Bank kehilangan kesempatan untuk fokus pada nasabah yang berniat mendepositokan dana mereka, sehingga potensi pendapatan dari deposito hilang.

**Asumsi**

Kita mendapatkan nilai rata-rata deposito dengan mengakses laporan keuangan bank Banco Comercial Portugues (BCP)

- Biaya campaign Marketing per Nasabah: (€2.15 + €1) = €3.15 
  - Gaji telemarketing di Portugal sekitar [1,676 Euro per bulan](https://www.statista.com/statistics/419498/average-annual-wages-portugal-y-on-y/ ).
  - Hari kerja dalam sebulan : 22 hari
  - Jam kerja dalam sehari : [8 jam](https://boundlesshq.com/guides/portugal/hours-of-work/)
  - Upah per jam = 1,676 Euro/bulan x (22 hari/ bulan x 8 jam/hari) = 9.53 Euro/jam
  - Rata-rata durasi telemarketing per orang sekitar 4.3 menit.
  - Gaji telemarketing per panggilan : €0,7 ≈ €1 
  - Rata-rata Durasi Panggilan: [4.3 menit](https://en.meo.pt/helplines)
  - Biaya per Menit untuk Panggilan Lokal : [€0.5](https://www.vodafone.pt/en/press-releases/2007/8/vodafone-portugal-reduces-and-simplifies-roaming-charges.html)
- Rata-rata Deposito per Nasabah yang Melakukan Deposito: [€2,400](https://www.ceicdata.com/en/portugal/banking-indicators/pt-deposit-accounts-per-1000-adults-commercial-banks)
- Rata-rata Bunga Deposito Bank Portugal/tahun (2010): [1.71%](https://ycharts.com/indicators/portugal_deposit_interest_rate)
- Rata-rata Bunga Pinjaman Bank Portugal/tahun (2010): [3.27%](https://take-profit.org/en/statistics/interest-rate/portugal/)

**Perhitungan Biaya**

**Biaya Marketing yang Terbuang Sia-sia (False Positive)**
- **Wasted Marketing Cost per Customer** = €3.15

**Kehilangan Potensi Pendapatan (False Negative)**
- **Potential Income Loss per Customer** = (Rata-rata Deposito per Nasabah × Rata-rata Bunga Pinjaman Bank Portugal/tahun) - (Rata-rata Deposito per Nasabah × Rata-rata Bunga Deposito Bank Portugal/tahun)
- **Potential Income Loss per Customer** = (€2,400 × 3.27%) - (€2,400 × 1.71%)
- **Potential Income Loss per Customer** = €78.48 - €41.04 = €37.44

Dari hasil perhitungan ini kita dapat melihat bahwa bank akan mengalami kerugian lebih besar saat kehilangan nasabah potensial. Oleh karena itu, **recall (True Positive Rate)** menjadi metrik utama yang harus dioptimalkan untuk meminimalkan kerugian akibat kesalahan prediksi dalam model machine learning.

Recall dihitung dengan rumus:

$$ \text{Recall} = \frac{\text{True Positives (TP)}}{\text{True Positives (TP)} + \text{False Negatives (FN)}} $$

- **True Positives (TP):** Kasus di mana model memprediksi 'Deposit' dan sebenarnya itu adalah 'Deposit'.
- **False Negatives (FN):** Kasus di mana model memprediksi 'No Deposit', tetapi sebenarnya itu adalah 'Deposit'.

- Contoh Perhitungan Biaya Berdasarkan Asumsi Jumlah Kesalahan

Misalkan kita memiliki:
- 1000 False Positives
- 200 False Negatives

**Total Biaya campaign yang Terbuang**:
$$ \text{Total Biaya Terbuang} = 1000 \times €3.15 = €3150 $$

**Total Kehilangan Potensi Pendapatan**:
$$ \text{Total Kehilangan Potensi Pendapatan} = 200 \times €37.44 = €7,488 $$

**Kesimpulan**
Dari perhitungan di atas, kita dapat melihat bahwa dampak biaya dari False Negatives jauh lebih signifikan dibandingkan dengan False Positives. Oleh karena itu, penting untuk meminimalkan kesalahan prediksi False Negatives guna mengurangi kehilangan potensi pendapatan yang besar.

### **2.5. Outliers**

```python
numerical = [var for var in data.columns if data[var].dtype != 'object']
categorical = [var for var in data.columns if data[var].dtype == 'object']

def calculate_outliers(data, column):
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = data[(data[column] < lower_bound) | (data[column] > upper_bound)]
    return outliers, lower_bound, upper_bound

outlier_data = []


for column in numerical:
    outlier_result, lower_bound, upper_bound = calculate_outliers(data, column)
    total_outlier = len(outlier_result)
    outlier_percentage = round(total_outlier / len(data[column]) * 100, 2)
    outlier_data.append([column, total_outlier, outlier_percentage, lower_bound, upper_bound])

outlier_df = pd.DataFrame(outlier_data, columns=["Column", "Total Outliers", "Percentage (%)", "Lower Bound", "Upper Bound"])

plt.figure(figsize=(12, 8))

for index, col in enumerate(numerical):
    # Plot boxplot
    plt.subplot(len(numerical) // 2 + len(numerical) % 2, 2, index + 1)
    sns.boxplot(x=data[col], color='lightblue')
    plt.xlabel(col, fontweight='bold')

plt.suptitle('Outliers of Numerical Variables', size=16, fontweight='bold')
plt.tight_layout()

display(outlier_df)
plt.show()
```

  1. MCAR (Missing Completely at Random): Data hilang secara acak, tanpa pola yang dapat dideteksi dan tidak tergantung pada nilai dari data lainnya atau data yang hilang itu sendiri. 

  2. MAR (Missing at Random) : Data hilang bukan secara acak, tetapi pola kehilangannya dapat dijelaskan oleh data lain yang diamati dalam dataset. 

  3. MNAR (Missing Not at Random): Data hilang tidak secara acak dan pola kehilangannya terkait langsung dengan nilai yang hilang itu sendiri atau faktor-faktor lain yang tidak diamati dalam dataset.

```python
# Function to calculate Cramér's V
def cramers_v(x, y):
    confusion_matrix = pd.crosstab(x, y)
    chi2 = chi2_contingency(confusion_matrix)[0]
    n = confusion_matrix.sum().sum()
    r, k = confusion_matrix.shape
    return np.sqrt(chi2 / (n * (min(k, r) - 1)))

# Example DataFrame with categorical variables
df = pd.DataFrame(demographic_data)

# Calculate the correlation matrix
cols = df.columns
n = len(cols)
corr_matrix = np.zeros((n, n))

for i in range(n):
    for j in range(n):
        corr_matrix[i, j] = cramers_v(df[cols[i]], df[cols[j]])

# Convert to demographic_dataFrame for better visualization
corr_df = pd.DataFrame(corr_matrix, index=cols, columns=cols)

# Plot the heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(corr_df, annot=True, cmap='coolwarm', center=0)
plt.title('Cramér\'s V Correlation Matrix')
plt.show()
```

### **4.2. Analisis Informasi Campaign Sebelumnya**

```python
numerical_features = ['duration', 'campaign', 'pdays', 'previous']
numerical_corr = data2[numerical_features + ['deposit']].corr()

categorical_features = ['contact', 'month', 'day_of_week', 'poutcome']
results = []

for feature in categorical_features:
    contingency_table = pd.crosstab(data2[feature], data2['deposit'])
    
    chi2, p, dof, expected = stats.chi2_contingency(contingency_table)
    
    if p < 0.05:
        significance = "Significant"
    else:
        significance = "Not Significant"
    
    results.append({
        'Feature': feature,
        'Chi2': chi2,
        'P-value': p,
        'P-value (round)': round(p, 3),
        'Significance': significance
    })

results_df = pd.DataFrame(results)

fig, ax= plt.subplots(figsize=(5, 4))

mask_num = np.triu(np.ones_like(numerical_corr, dtype=bool))
sns.heatmap(numerical_corr, annot=True, cmap='coolwarm', mask=mask_num)
ax.set_title('Correlation Matrix for Numerical Features')

plt.tight_layout()
plt.show()
results_df
```

```python
def calculate_vif(data):
    vif = pd.DataFrame()
    vif['Feature'] = data.columns
    vif['VIF'] = [variance_inflation_factor(data.dropna(), i) for i in range(len(data.columns))]

    return vif.sort_values('VIF', ascending=False)
```

Kita akan melakukan pengujian Multicolinearity. Apa itu multicolinearity dan mengapa pengujian multicolinearity penting? Multikolinearity adalah terjadinya interkorelasi tinggi di antara dua atau lebih variabel independen dalam model regresi.  High Multicolinearity dapat menyebabkan interval kepercayaan (confidence interval) yang lebih lebar yang menghasilkan probabilitas yang kurang dapat diandalkan dalam hal pengaruh variabel independen dalam suatu model.

Dalam pengujian kali ini kita akan melakukan pengujian berulang hingga diperoleh nilai VIF < 10.

```python
vif_bankcamp = calculate_vif(data_copy[numerical])
dropped_columns = []

for i in range(len(vif_bankcamp)):
    if vif_bankcamp.iloc[0, -1] < 10:
        break
    else:
        dropped_columns.append(vif_bankcamp.iloc[0, 0])
        vif_hoetl = calculate_vif(data_copy[numerical].drop(columns=dropped_columns))
```

## Handling Outliers

```python
# Cek persentase outlier untuk setiap numerical feature
def calculate_outliers_1(data, column):
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = data[(data[column] < lower_bound) | (data[column] > upper_bound)]
    return outliers, lower_bound, upper_bound

outlier_data = []

numerical = [var for var in data_copy.columns if data_copy[var].dtypes != 'object']

for column in numerical:
    if column == 'deposit' :
        continue
    outlier_result, lower_bound, upper_bound = calculate_outliers_1(data_copy, column)
    total_outlier = len(outlier_result)
    outlier_percentage = round(total_outlier / len(data_copy[column]) * 100, 2)
    outlier_data.append([column, total_outlier, outlier_percentage, lower_bound, upper_bound])

outlier_df = pd.DataFrame(outlier_data, columns=["Column", "Total Outliers", "Percentage (%)", "Lower Bound", "Upper Bound"])

display(outlier_df)
```

Terdapat 2 kategori outlier yang ada dalam dataset ini :

- **Extreme Outlier** (>10%) = `default` dan `previous`

| `learning_rate`           | [0.01, 0.05, 0.1]       | Tingkat pembelajaran yang dipilih untuk mengontrol kontribusi setiap estimator terhadap model akhir. Memilih nilai yang rendah untuk pembelajaran yang lebih hati-hati. |

### **Decision Tree**

| Parameter                  | Range                          | Deskripsi                                                                                                                                                   |
|----------------------------|--------------------------------|----------------------------------------------------------------------------------------------------------------------------------------------------------|
| `max_depth`                | [None, 5, 10, 15, 20, 25]      | Kedalaman maksimum dari setiap pohon. Memilih nilai `None` untuk mengijinkan pohon tumbuh sampai tidak terbatas, atau nilai lain untuk mengontrol kompleksitas model. |
| `min_samples_split`        | [2, 5, 10, 15, 20]             | Jumlah sampel minimum yang diperlukan untuk membagi node internal. Memilih nilai yang lebih tinggi untuk mencegah overfitting.                             |
| `min_samples_leaf`         | [1, 2, 4, 6, 10]               | Jumlah sampel minimum yang diperlukan untuk menjadi leaf (daun) pada pohon. Memilih nilai yang lebih tinggi untuk mencegah overfitting.                      |

| `criterion`                | ['gini', 'entropy']            | Fungsi untuk mengukur kualitas split. Memilih antara gini impurity atau entropy based on information gain.                                                 |
| `ccp_alpha`                | [0.0, 0.01, 0.02, 0.05, 0.1]   | Parameter pruning cost complexity. Memilih nilai yang lebih tinggi untuk pruning yang lebih agresif dan menghindari overfitting.                           |

| Parameter | Range | Description |

```python
def metrics_calculator(clf, X_test, y_test, model_name):
    
    y_pred = clf.predict(X_test)
    result = pd.DataFrame(data=[accuracy_score(y_test, y_pred),
                                precision_score(y_test, y_pred, pos_label=1),
                                recall_score(y_test, y_pred, pos_label=1),
                                f1_score(y_test, y_pred, pos_label=1),
                                roc_auc_score(y_test, clf.predict_proba(X_test)[:,1])],
                          index=['Accuracy','Precision (Class 1)','Recall (Class 1)','F1-score (Class 1)','AUC (Class 1)'],
                          columns = [model_name])
    
    result = (result * 100).round(2).astype(str) + '%'                            
    return result


def evaluation_visualization(clf, X_train, X_test, y_train, y_test, model_name):
    '''
    Fungsi ini bertujuan hanya untuk meanampilkan Confusiion Metrics, AUC curve, dan PR curve
    '''
    sns.set(font_scale=1.2)
    
    # Create figure and subplots 
    fig, (ax1, ax2, ax3, ax4) = plt.subplots(1, 4, figsize=(20, 5), dpi=100, gridspec_kw={'width_ratios': [2, 2, 2, 1]}, facecolor='lightsteelblue')
    
    # Define a colormap
    royalblue = LinearSegmentedColormap.from_list('crimson', [(0, (1,1,1)), (1, (0.25,0.41,0.88))])
    royalblue_r = royalblue.reversed()

    # Plot confusion matrix 
    ConfusionMatrixDisplay.from_estimator(clf, X_test, y_test, colorbar=False, cmap=royalblue_r, ax=ax1)
    ax1.set_title('Confusion Matrix for Test Data')                                     
    ax1.grid(False)
    
    # Plot AUC-ROC 
    RocCurveDisplay.from_estimator(clf, X_test, y_test, ax=ax2)
    ax2.set_xlabel('False Positive Rate')
    ax2.set_ylabel('True Positive Rate')
    ax2.set_title('ROC Curve for Test Data (Positive label: 1)')
    
    # Plot Precision-Recall curve
    PrecisionRecallDisplay.from_estimator(clf, X_test, y_test, ax=ax3)
    ax3.set_xlabel('Recall')
    ax3.set_ylabel('Precision')
    ax3.set_title('Precision-Recall Curve for Test Data (Positive label: 1)')

    # Rekap hasil evaluasi
    result = metrics_calculator(clf, X_test, y_test, model_name)
    table = ax4.table(cellText=result.values, colLabels=result.columns, rowLabels=result.index, loc='center')
    table.scale(0.6, 2)
    table.set_fontsize(12)
    ax4.axis('tight')
    ax4.axis('off')
    # Modify color 
    for key, cell in table.get_celld().items():
        if key[0] == 0:
            cell.set_color('moccasin')

    # Add the title for the entire figure
    fig.suptitle(model_name, fontsize=16, fontweight='bold')
    
    plt.tight_layout(rect=[0, 0, 1, 0.95])  
    plt.show()
```

```python
evaluation_visualization(random_search_lgbm, X_train, X_test, y_train, y_test, 'Light GBM')
evaluation_visualization(random_search_xgb, X_train, X_test, y_train, y_test, 'Extreme Gradient Boost')
evaluation_visualization(random_search_tree, X_train, X_test, y_train, y_test, 'Decision Tree')
evaluation_visualization(grid_search_ada, X_train, X_test, y_train, y_test, 'ADA Boost')
```

Confusion Matrix menunjukkan bahwa meskipun ada banyak false positives (4296), jumlah true positives (856) sangat tinggi.

- **Decision Tree** menunjukkan precision yang lebih tinggi namun recall sangat rendah, dengan ROC AUC yang lebih rendah (0.65), menandakan kemampuan model untuk membedakan antara kelas yang berbeda kurang baik. Confusion Matrix menunjukkan banyak false negatives (645) dan false positives (327).

LightGBM (Light Gradient Boosting Machine) adalah algoritma peningkatan gradien berbasis decision tree yang sangat efisien dan cepat. Model ini membagi data menggunakan metode leaf-wise, yang dapat menghasilkan akurasi yang lebih tinggi dibandingkan metode level-wise pada sejumlah dataset.
```python
# menyimpan algoritma model terbaik dari Gradient Boost sebagai best model
best_model = Pipeline([
    ('transformer', transformer),
    ('scaler', scaler),
    ('model', LGBMClassifier(scale_pos_weight=30,num_leaves=31,n_estimators=100 , max_depth=10, max_bin=275,learning_rate=0.05))
])

best_model = random_search_lgbm.best_estimator_
y_pred_best_model = best_model.predict(X_test)
recall = recall_score(y_test, y_pred_best_model)
print("Recall Score on Test Data:", recall)
```

Recall score 95,6% berarti bahwa model berhasil mengidentifikasi 95% dari semua pelanggan yang sebenarnya melakukan deposito berdasarkan campaign bank.

Recall dihitung dengan rumus:

$$ \text{Recall} = \frac{\text{True Positives (TP)}}{\text{True Positives (TP)} + \text{False Negatives (FN)}} $$

- **True Positives (TP):** Kasus di mana model memprediksi 'Deposito' dan sebenarnya nasabah memamng melakukan 'Deposito'.
- **False Negatives (FN):** Kasus di mana model memprediksi 'Tidak Deposito', tetapi sebenarnya nasabah melakukan 'Deposito'.

Jadi, recall 95.6% menunjukkan bahwa dari semua pelanggan yang benar-benar melakukan deposito, 95% di antaranya berhasil diidentifikasi oleh model sebagai melakukan deposito. Namun, 5% sisanya tidak teridentifikasi (model menganggap mereka tidak melakukan deposito).

```python
def proba_range(x):
    if x < 0.1:
        proba_range = '0.0 - 0.1'
    elif 0.1 <= x < 0.3:
        proba_range = '0.1 - 0.3'
    elif 0.3 <= x < 0.5:
        proba_range = '0.3 - 0.5'
    elif 0.5 <= x < 0.6:
        proba_range = '0.5 - 0.6'
    elif 0.6 <= x < 0.7:
        proba_range = '0.6 - 0.7'
    elif 0.7 <= x < 0.8:
        proba_range = '0.7 - 0.8'
    elif 0.8 <= x < 0.9:
        proba_range = '0.8 - 0.9'
    else:
        proba_range = '0.9 - 1.0'
        
    return proba_range

data_proba.insert(1, 'proba_range',value ='')

data_proba['proba_range'] = data_proba['Proba'].apply(proba_range)
```

```python
# Mengelompokkan data berdasarkan rentang probabilitas dan menghitung ukuran masing-masing grup
proba_set = data_proba.groupby(['proba_range', 'y_test']).size().unstack().reset_index()

# Mengganti nama kolom untuk meningkatkan keterbacaan
proba_set = proba_set.rename(columns={0: 'no', 1: 'yes'})

# Menambahkan kolom total dan actual probability
proba_set['total'] = proba_set['no'] + proba_set['yes']
proba_set['actual probability'] = (proba_set['yes'] / proba_set['total']).round(2)

# Menambahkan kolom predicted probability
predicted_probs = [0.05, 0.2, 0.4, 0.55, 0.65, 0.75, 0.85, 0.95]
proba_set['predicted probability'] = predicted_probs[:len(proba_set)]

# Mengurutkan DataFrame berdasarkan rentang probabilitas
proba_set = proba_set.sort_values(by='proba_range', ascending=False)

proba_set
```

### Model Limitation

Ketika menggunakan model machine learning seperti LightGBM, penting untuk berhati-hati dalam melakukan interpretasi hasil terutama di luar rentang data yang digunakan dalam pelatihan model. Berikut adalah beberapa batasan spesifik dari model ini berdasarkan rentang variabel yang digunakan:

**1. Batasan Rentang Usia (`age`):**
   - **Deskripsi**: Model ini dilatih menggunakan data dengan rentang usia nasabah antara 17 hingga 100 tahun.
   - **Implikasi**: Prediksi model mungkin tidak akurat untuk nasabah yang berada di luar rentang usia ini, misalnya nasabah yang lebih muda dari 17 tahun atau lebih tua dari 100 tahun. Oleh karena itu, interpretasi hasil untuk nasabah di luar rentang usia ini harus dilakukan dengan hati-hati.

**2. Batasan Jumlah Kampanye (`campaign`):**
   - **Deskripsi**: Data yang digunakan dalam pelatihan model memiliki jumlah kampanye yang berkisar antara 1 hingga 56 kali.
   - **Implikasi**: Model mungkin tidak memberikan prediksi yang akurat untuk nasabah yang telah dihubungi lebih dari 56 kali atau kurang dari 1 kali. Oleh karena itu, penting untuk mempertimbangkan batasan ini saat menilai prediksi untuk nasabah dengan jumlah kampanye yang ekstrem.

**3. Batasan Jumlah Kontak Sebelumnya (`previous`):**
   - **Deskripsi**: Rentang data untuk jumlah kontak sebelumnya dengan nasabah adalah antara 0 hingga 7 kali.
   - **Implikasi**: Model mungkin tidak valid untuk nasabah yang memiliki jumlah kontak sebelumnya lebih dari 7 kali. Interpretasi hasil untuk kasus seperti ini harus dilakukan dengan hati-hati.

### Rekomendasi untuk Mengatasi Limitasi

1. **Penyegaran Data secara Berkala**: 
   - Secara berkala perbarui data pelatihan model dengan data terbaru untuk mencakup perubahan dalam variabel-variabel penting. Hal ini membantu model tetap relevan dan akurat.

2. **Ekstrapolasi dengan Hati-Hati**: 
   - Hindari melakukan prediksi untuk data di luar rentang variabel yang telah diamati selama pelatihan model. Jika ekstrapolasi tidak dapat dihindari, gunakan teknik-teknik validasi tambahan untuk memastikan keakuratan prediksi.

3. **Monitoring Kinerja Model**: 
   - Implementasikan sistem monitoring untuk memantau kinerja model secara terus-menerus. Jika ditemukan penurunan performa, lakukan retraining model dengan data yang lebih relevan.

4. **Penerapan Teknik Feature Engineering**: 
   - Gunakan teknik feature engineering untuk memastikan bahwa variabel yang digunakan dalam model benar-benar relevan dan mencerminkan kondisi terkini. Hal ini dapat membantu meningkatkan akurasi dan validitas prediksi.

5. **Analisis Sensitivitas**: 
   - Lakukan analisis sensitivitas untuk memahami bagaimana perubahan dalam variabel-variabel input mempengaruhi prediksi model. Ini membantu dalam mengidentifikasi variabel mana yang paling kritis dan perlu diperhatikan.
